```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_colwidth', 100)

from statsmodels.tsa.stattools import grangercausalitytests
```

# Import Data

```python
stock_all = pd.read_csv('./stock_all.csv')
stock_all['date'] = pd.to_datetime(stock_all['Date'], errors='coerce')
stock_all.set_index('date', inplace=True)
stock_all.drop('Date', axis=1, inplace=True)
```

```python
stock_all_mean = pd.DataFrame(stock_all['close_mean'])
```

```python
sentiment_all = pd.read_csv('./twitter_sentiment_daily_all.csv')
sentiment_all['date'] = pd.to_datetime(sentiment_all['Date'], errors='coerce')
sentiment_all.set_index('date', inplace=True)
sentiment_all.drop('Date', axis=1, inplace=True)
```

```python
sentiment_all
```

# Merge Data 

```python
data = sentiment_all.copy()
```

```python
data = pd.merge(data, stock_all_mean, how='left', left_on=['date'], right_on=['date'] )
```

```python
data.rename(columns={'neg':'neg_sentiment','pos':'pos_sentiment','compound':'compound_sentiment','close_mean':'closing_price'}, inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data
```

# Lag Analysis

```python
max_lag = 1000
```

```python
def plot_cross_correlation(series_1, series_2, max_lag):
    fig, ax1 = plt.subplots(figsize = (18,5))
    ax1.xcorr(series_1, series_2, usevlines = True, maxlags = max_lag, normed=True)
    ax1.set_ylim(0,1)
    ax1.grid(True)
    plt.show()
```

```python
series_1 = data['pos_sentiment'].to_numpy().flatten()
series_2 = data['closing_price'].to_numpy().flatten()
plot_cross_correlation(series_1, series_2, max_lag)
```

```python
series_1 = data['neg_sentiment'].to_numpy().flatten()
series_2 = data['closing_price'].to_numpy().flatten()
plot_cross_correlation(series_1, series_2, max_lag)
```

```python
series_1 = data['compound_sentiment'].to_numpy().flatten()
series_2 = data['closing_price'].to_numpy().flatten()
plot_cross_correlation(series_1, series_2, max_lag)
```

```python
data
```

```python
series_1 = data['compound_sentiment']['2020-01-01':'2021-04-01'].to_numpy().flatten()
series_2 = data['closing_price']['2020-01-01':'2021-04-01'].to_numpy().flatten()
plot_cross_correlation(series_1, series_2, max_lag)
```

# Granger Causality 

```python
data.columns
```

```python
input_data = np.asarray(data[['compound_sentiment', 'closing_price']])
maxlag = 10
grangercausalitytests(input_data, maxlag, addconst=True, verbose=True)
```

```python
input_data = np.asarray(data[['pos_sentiment', 'closing_price']])
maxlag = 10
grangercausalitytests(input_data, maxlag, addconst = True, verbose=True)
```
